```python
import urllib.request
import urllib.parse
import requests
import re

headers = {'User-Agent' : "sdf"}
url=input("Enter Url : ")

response = requests.get(url, headers= headers)
html=response.text
temp_index=html.find("data-old-hires" )
imgstIndex=html.find("https://images",temp_index)
imgfnIndex=html.find("jpg", imgstIndex)+3
imgurl=html[imgstIndex:imgfnIndex]

print(imgurl)


url = html[imgstIndex:imgfnIndex]
filenamestIndex=html.find("<meta name=\"title\" content=\"")+28
filenamefnIndex=html.find("\"",filenamestIndex)
savename=html[filenamestIndex:filenamefnIndex]
savename=re.sub("[\/:*?\"<>|]","",savename)


mem = requests.get(url).content
with open(savename+'.jpg', "wb") as f:
    f.write(mem)
    print("저장 완료")
    
f=open(savename+".txt",'w', encoding='UTF-8')

if(html.find("Technische Details")!=-1):
    html_selection=html[html.find("Technische Details"): html.find("Zusätzliche Produktinformationen")]
else:
    html_selection=html[html.find("Produktdetails"): html.find("Zusätzliche Produktinformationen")]

productInfo=''
productInfo+=html[filenamestIndex:filenamefnIndex]
productInfo+='\n\n'

productInfo+="Preis: "
st=html.find("<span id=\"priceblock_ourprice\" class=\"a-size-medium a-color-price priceBlockBuyingPriceString\">")+95
fn=html.find("</span>", st)
productInfo+=html[st:fn]
productInfo+='\n\n'


labelStIndex=0
labelFnIndex=0
ValueStIndex=0
ValueFnIndex=0

if ("<td class=\"label\">" in html):
    while True:
        labelStIndex=html_selection.find("<td class=\"label\">", ValueFnIndex)+18
        labelFnIndex=html_selection.find("</td>", labelStIndex)
        ValueStIndex=html_selection.find("<td class=\"value\">", labelStIndex)+18
        ValueFnIndex=html_selection.find("</td></tr>", ValueStIndex)
        productInfo+=html_selection[labelStIndex:labelFnIndex]
        productInfo+=": "
        productInfo+=html_selection[ValueStIndex:ValueFnIndex]
        productInfo+='\n'
        if(html_selection.find("<td class=\"label\">", ValueStIndex)==-1):
            break
            

elif ("<td class=\"label\">" not in html):
    labelStIndex=0
    labelFnIndex=0
    ValueStIndex=0
    ValueFnIndex=0
    while True:
        labelStIndex=html.find("<li><b>",ValueFnIndex)+7
        labelFnIndex=html.find("</b>", labelStIndex)
        ValueStIndex=labelFnIndex+4
        ValueFnIndex=html.find("</li>",ValueStIndex)
        productInfo+=html[labelStIndex:labelFnIndex]+' '
        productInfo+=html[ValueStIndex:ValueFnIndex]+'\n'
        if(html.find("<li><b>", ValueFnIndex)==-1):
            break
        


f.write(productInfo)

f.close()
```

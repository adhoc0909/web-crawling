```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd

disp_category_number=str(input("카테고리 번호(ex: 100040000600001) disp_cat_no= "))
big_category_number=disp_category_number[:-2]

while True:
    file_name=str(input("파일이름(ex: bar_code_icecream.csv) :"))
    if file_name[-4:]!='.csv':
        print(".csv를 붙여야 합니다.")
        continue
    else:
        break

base_url="http://www.lottesuper.co.kr/handler/category/CategoryList-Start?disp_cat_no={}{}&pageNo=1&rowsPerPage=80&sort=0&view_type=0&benefits=&tabPageNo=0&iconType=C".format(big_category_number,'01')
driver=webdriver.Firefox(capabilities=None, executable_path='C:\\Users\\admin\\Downloads\\geckodriver-v0.26.0-win64\\geckodriver')
driver.implicitly_wait(3)

#웹페이지 로드
driver.get(base_url)

error_cnt=0

end_cate=int(input("number of categories: "))

for category in range(1, end_cate+1):
    if(category<10):
        category_number='0'+str(category)
    else:
        category_number=str(category)
        
    driver.get("http://www.lottesuper.co.kr/handler/category/CategoryList-Start?disp_cat_no={}{}&pageNo={}&rowsPerPage=80&sort=0&view_type=0&benefits=&tabPageNo=0&iconType=C".format(big_category_number, category_number, 1))
    if(len(driver.find_elements_by_xpath("//*[@class='num_box']"))==0):
        end_page=1
    else:
        end_page=int(driver.find_elements_by_xpath("//*[@class='num_box']")[-1].text)
    for page in range(1, end_page+1):
        driver.get("http://www.lottesuper.co.kr/handler/category/CategoryList-Start?disp_cat_no={}{}&pageNo={}&rowsPerPage=80&sort=0&view_type=0&benefits=&tabPageNo=0&iconType=C".format(big_category_number, category_number, page))
      
        for product in range(1,81):
            try:
                driver.find_element_by_xpath("/html/body/div[2]/div[3]/form/fieldset/div[2]/div[2]/div/div[4]/div/div[2]/ul/li[{}]/em/a".format(product)).click()
                title=driver.find_element_by_xpath("//*[@class='goodsTit']").text
                code=driver.find_element_by_xpath("//*[@class='productInfo last']").text.split(' / ')[-1]
                code+='\''
                a=pd.DataFrame({'title' : title, 'code' : code}, index=['-'])
                a.to_csv(file_name, mode='a', index=False, header=False)
                print("csv 저장 완료", title, "/", code)
                driver.back()
                print("카테고리: (", category,"/", end_cate, ") 페이지: (", page, "/", end_page, ") &", str(product)+"번째 제품 완료.")
            except PermissionError:
                error_cnt+=1
                print(error_cnt, "접근 오류발생", "카테고리:",category, "페이지:",page, "상품:",product)
                continue
            except: #이 예외처리에 의한 오류가 떠도 상품 저장에 지장이 없음. 한 페이지에 상품이 80개가 안될 때 뜨는 오류임.
                error_cnt+=1
                print(error_cnt,"번째 오류", "카테고리:",category, "페이지:",page, "상품:",product)
                continue
                
                
```
